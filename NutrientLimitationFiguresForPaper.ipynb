```python
ProdNorm = IntPP/mean(IntPP,0)
ProdNormControl = IntPPControl/mean(IntPPControl,0)
ProdNormTrans = IntPPTrans/mean(IntPPTrans,0)
ProdNormFrugal = IntPPFrugal/mean(IntPPFrugal,0)
ProdNormGrowthRate = IntPPGrowthRate/mean(IntPPGrowthRate,0)
```

```python
# Here we will construct "mean-limitation matrices"

NLimMatLowLight = (mean(SmNLim*ProdNorm,0))
PLimMatLowLight = (mean(SmPLim*ProdNorm,0))
FeLimMatLowLight = (mean(SmFeLim*ProdNorm,0))


# We want to define categories: N-Lim, P-Lim, Fe-Lim, NP Colim, NFe Colim, PFe Colim

# N Lim is NLim/PLim and NLim/FeLim both < Threshold (or could also base it on the log)
# logNLim - logPLim < threshold and logNLim -logFeLim < threshold

# Start with threshold of -0.3

ColimThreshold = 0.1
RepThreshold = 0.8


RepleteMat = (NLimMatLowLight > (RepThreshold)) * (PLimMatLowLight > (RepThreshold)) * (FeLimMatLowLight > (RepThreshold))



NLimMat = (NLimMatLowLight - PLimMatLowLight < 0) * (NLimMatLowLight - FeLimMatLowLight < 0) 
PLimMat = (PLimMatLowLight - NLimMatLowLight < 0) * (PLimMatLowLight - FeLimMatLowLight < 0) 
FeLimMat = (FeLimMatLowLight - NLimMatLowLight < 0) * (FeLimMatLowLight - PLimMatLowLight < 0) 



FullNLimMat = (NLimMatLowLight - PLimMatLowLight < -ColimThreshold) * (NLimMatLowLight - FeLimMatLowLight < -ColimThreshold) 
FullPLimMat = (PLimMatLowLight - NLimMatLowLight < -ColimThreshold) * (PLimMatLowLight - FeLimMatLowLight < -ColimThreshold) 
FullFeLimMat = (FeLimMatLowLight - NLimMatLowLight < -ColimThreshold) * (FeLimMatLowLight - PLimMatLowLight < -ColimThreshold) 

WeakNLimMat = NLimMat*(1-FullNLimMat)
WeakPLimMat = PLimMat*(1-FullPLimMat)
WeakFeLimMat = FeLimMat*(1-FullFeLimMat)


LimPatternSm = FullNLimMat+2*WeakNLimMat+3*WeakPLimMat+4*FullPLimMat+5*WeakFeLimMat+6*FullFeLimMat
                        
    


fig = plt.figure(figsize=(10,4))
projection = ct.crs.Mollweide(central_longitude=200)
proj2 = ct.crs.PlateCarree()



ax = fig.add_subplot(1,1,1,projection=projection)
ax.coastlines(resolution="110m")



ax.pcolormesh(lon,lat,LimPatternSm,transform=proj2,cmap="turbo")

#colors = [ im.cmap(im.norm(value)) for value in range(1,7)]
# create a patch (proxy artist) for every color 
#labelVec = ["N-Lim","Weak N-Lim","Weak P-Lim", "P-Lim","Weak Fe-Lim","Fe-Lim"]

#patches = [ matplotlib.patches.Patch(color=colors[i], label=labelVec[i] ) for i in range(len(colors)) ]
# put those patched as legend-handles into the legend
#ax.legend(handles=patches, bbox_to_anchor=(.025, 0.85), loc=2, borderaxespad=0.,fontsize=20 )

#ax.set_title("GM N:P",fontsize=24)

#plt.savefig("NutLimATOM.png",bbox_inches="tight",dpi=300)
```
